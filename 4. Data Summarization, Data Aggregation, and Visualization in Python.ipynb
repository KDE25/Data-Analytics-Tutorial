## Basic Aggregation

```python
# Mean Aggregation based on a column
df_summer.groupby(['Country'])[['medal_rank']].mean()
```

```python
# Median Aggregation based on a column
df_summer.groupby(['Country'])[['medal_rank']].median()
```

```python
# Mean Aggregation based on a column & Sorted
df_summer.groupby(['Country'])[['medal_rank']].mean().sort_values(by=['medal_rank'],ascending=False)
```

## Multilevel Aggregation

```python
# Mean Aggregation based on several columns
df_summer.groupby(['Discipline','Country'])[['medal_rank']].mean()
```

```python
# Mean Aggregation based on several columns
df_summer.groupby(['Discipline','Country'])[['medal_rank']].agg(['mean','min','max'])
```

```python
# Counting the number of occurences

pd.DataFrame(df_summer.groupby(['Discipline','Country','Medal']).size())
```

```python
pd.DataFrame(df_summer.groupby(['Discipline','Country','Medal']).size())
```

Another way to do aggregation is using `pivot_table`. The only difference between `groupby` and `pivot_table` is the output shape.

```python
df_summer.pivot_table(index=['Discipline','Country'], columns='Medal', values='medal_rank', aggfunc=len, fill_value=0)
```

### Creating Basic Grid

Sebagai pendahuluan, mari kita lihat kode dasar membuat 'alas' dari grafik yang akan kita buat.

```python
fig = plt.figure()  # membuat 'figure' baru
ax = plt.axes() # membuat sumbu (x dan y)
plt.grid() # membuat petak-petak atau gridlines
plt.show() # menampilkan plot
```

Dengan mengganti parameter color pada plt.grid(), kita bisa mengganti warna gridlines menjadi warna sesuka hati kita.


```python
fig = plt.figure()
ax = plt.axes()
plt.grid(color='red')
plt.show()
```

Terdapat beberapa cara menjabarkan warna di matplotlib.

* Warna-warna dasar seperti color = 'red', color = 'purple', color = 'blue', dan seterusnya
* Warna RGB dengan kode #RRGGBB. Contoh: color = #824394. Bagaimana cara tahu kode RGB dari warna yang kita inginkan? https://imagecolorpicker.com/color-code/2596be. Website ini mempermudah kita memilih warna yang kita inginkan, dan kemudian mereka akan memberitahukan kita kode RGB dari warna tersebut
* Warna-warna yang tidak umum seperti 'AntiqueWhite', 'Moccasin' dst dapat dilihat dari daftar di sini: https://www.w3schools.com/colors/colors_names.asp


Kita juga bisa mengatur tebal tipisnya garis, dan juga style dari garis itu sendiri

```python
fig = plt.figure()
ax = plt.axes()
plt.grid(visible=True, color='black',
         alpha=0.5, # mengatur opacity garis
         linestyle='-.', # mengatur 'tekstur' garis
         linewidth=2.4) # mengatur ketebalan garis
```

Terdapat dua macam grid di Matplotlib, Major dan Minor. Mari kita coba lihat perbedaannya

```python
x = np.linspace(0, 5, 1000)

fig = plt.figure()
ax = plt.axes()
plt.plot(x, np.sin(x), color='Indigo', linestyle='--', linewidth=3)

# Major Grid berwarna biru
plt.grid(color='blue', which = 'major', alpha=0.2, linestyle='-.', linewidth=2)
# Minor Grid berwarna merah
plt.minorticks_on()
plt.grid(color = 'red', which = 'minor', alpha=0.4, linestyle='-', linewidth=1)

plt.show()
```

Sebagai contoh, grafik di atas cukup menggambarkan perbedaan Major dan Minor grid. Namun, secara estetika, tidak disarankan membuat grafik dengan grid minor berwarna 'ngejreng' dan terlalu mengganggu mata seperti di atas.

```python
x = np.linspace(0, 5, 1000)

fig = plt.figure()
ax = plt.axes()
plt.plot(x, np.sin(x), color='Indigo', linestyle='--', linewidth=3)

# Major Grid berwarna biru
plt.grid(color='blue', which = 'major', alpha=0.2, linestyle='-.', linewidth=2)
# Minor Grid berwarna abu-abu
plt.minorticks_on()
plt.grid(color = 'grey', which = 'minor', alpha=0.3, linestyle='-', linewidth=0.3)

plt.show()
```

#### Multiple Lines in One Area

Selain grafik sin, mari kita tambahkan juga grafik cos.

```python
x = np.linspace(-2, 10, 1000)

fig = plt.figure()
ax = plt.axes()

# Menambahkan grafik sinus dan cosinus
line1, = plt.plot(x, 2*np.sin(x), color='Indigo', linestyle='--', linewidth=3, label = 'y = 2sin(x)')
line2, = plt.plot(x, np.cos(x), color = 'darkgreen', linewidth = 3, label = 'y = cos(x)')

# Memberikan perintah agar muncul legenda pada plot kita
plt.legend(handles=[line1, line2], loc='best')

# Major Grid berwarna biru
plt.grid(color='blue', which = 'major', alpha=0.2, linestyle='-.', linewidth=2)
# Minor Grid berwarna abu-abu
plt.minorticks_on()
plt.grid(color = 'grey', which = 'minor', alpha=0.3, linestyle='-', linewidth=0.3)

ax.set_xlabel('Sumbu-x')
ax.set_ylabel('Sumbu-y')

# Set batas batas x dan y
plt.xlim(-2,10)
plt.ylim(-2.5, 2.5)

plt.suptitle('Grafik 2*Sin(x)', fontsize = 'x-large', ha = 'left', x = 0.1275)
plt.title('Nilai x dari -2 sampai 10', fontsize = 'medium', ha = 'left', x = 0)

# Rotate x-axis
plt.xticks(rotation=0)

plt.show()
```

### Creating Subplots

Terkadang kita ingin membuat beberapa plot terpisah dalam sebuah figure. Berikut cara membagi satu figure menjadi beberapa subplots.

```python
## Buat data x dan y yang baru
x1 = np.linspace(-1, 5, 1000)

y1 = np.sin(x1)
y2 = np.sin(x1**2)
y3 = np.cos(x1)
y4 = np.cos(x1**2)
```

Cara membuat subplot adalah dengan mengaturnya pada baris pertama coding kita (baris fig dan ax)

```python
fig, (ax1, ax2) = plt.subplots(2)
fig.suptitle('Vertically stacked subplots')
ax1.plot(x1, y1)
ax2.plot(x1, y2)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1,2)
fig.suptitle('Horizontally stacked subplots')
ax1.plot(x1, y1)
ax2.plot(x1, y2)
plt.show()
```

#### Review

1. Gambar Density Plot (KDE) semua distribusi ini pada satu plot, kemudian menggunakan faceting
2. Gambar Boxplot semua distribusi ini pada satu plot
3. Customize semua plot yang dibuat (beri judul, warna, dst)

```python
random_dists = ['Normal(1, 1)', 'Lognormal(1, 1)', 'Exp(1)', 'Gumbel(6, 4)',
                'Triangular(2, 9, 11)', 'Poisson(1)']
N = 500

norm = np.random.normal(1, 1, N)
logn = np.random.lognormal(1, 1, N)
expo = np.random.exponential(1, N)
gumb = np.random.gumbel(6, 4, N)
tria = np.random.triangular(2, 9, 11, N)
pois = np.random.poisson(1, N)
```

```python
sns.kdeplot(logn)
plt.show()
```

```python
### Density Plot - All at Once

fig = plt.figure(figsize = (10,10))
ax = plt.axes()

hist_norm = sns.kdeplot(norm, color = 'blue', label = 'Normal(1, 1)')
hist_logn = sns.kdeplot(logn, color = 'lightblue', label = 'Lognormal(1, 1)')
hist_expo = sns.kdeplot(expo, color = 'red', label = 'Exp(1)')
hist_gumb = sns.kdeplot(gumb, color = 'pink', label = 'Gumbel(6, 4)')
hist_tria = sns.kdeplot(tria, color = 'purple', label = 'Triangular(2, 9, 11)')
hist_pois = sns.kdeplot(pois, color = 'green', label = 'Poisson(1)')

plt.legend()

plt.grid(color = 'blue', which = 'major', alpha = 0.1, linestyle = '-.', linewidth=2)
plt.minorticks_on()
plt.grid(color = 'grey', which = 'minor', alpha = 0.2, linestyle = '-', linewidth = 0.3)

ax.set_xlabel('Sumbu-x')
ax.set_ylabel('Sumbu-y')

plt.show()
```

```python
### Density Plot - Faceting

fig, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2,3, figsize = (15,8), sharey = False)

# Draw the kde plots
hist_norm = sns.kdeplot(norm, color = 'blue', label = 'Normal(1, 1)', ax = ax1)
hist_logn = sns.kdeplot(logn, color = 'lightblue', label = 'Lognormal(1, 1)', ax = ax2)
hist_expo = sns.kdeplot(expo, color = 'red', label = 'Exp(1)', ax = ax3)
hist_gumb = sns.kdeplot(gumb, color = 'pink', label = 'Gumbel(6, 4)', ax = ax4)
hist_tria = sns.kdeplot(tria, color = 'purple', label = 'Triangular(2, 9, 11)', ax = ax5)
hist_pois = sns.kdeplot(pois, color = 'green', label = 'Poisson(1)', ax = ax6)

# Set axis label
ax1.set_ylabel('Density')
ax4.set_ylabel('Density')

ax2.set_ylabel('')
ax3.set_ylabel('')
ax5.set_ylabel('')
ax6.set_ylabel('')

# Set Title for each plots
ax1.set_title('Normal(1,1)')
ax2.set_title('Lognormal(1,1)')
ax3.set_title('Exp(1)')
ax4.set_title('Gumbel(6,4)')
ax5.set_title('Triangular(2,9,11)')
ax6.set_title('Poisson(1)')

fig.suptitle('Density Plots of Multiple Distributions', fontsize = 'xx-large')
plt.show()
```
