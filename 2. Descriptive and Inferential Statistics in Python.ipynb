### Mendefinisikan fungsi dengan Python.

- Fungsi blok dimulai dengan  kata kunci **def** diikuti oleh nama fungsi dan tanda kurung (())sebagai input.
- Argumen adalah input dari luar fungsi untuk dapat diproses oleh fungsi.
- Return mengembalikan nilai dari proses yang dilakukan sebagai output.

````python
# fungsi tanpa input
def greet():
    return "Halo!"

# fungsi dengan satu input
def luasLingkaran(radius):
    return 3.14 * radius ** 2

# fungsi dengan lebih dari satu input
def luasBalok(panjang, lebar, tinggi):
    return panjang * lebar * tinggi

# penerapan star assignment untuk "menangkap" kumpulan kumpulan input
def jumlah(*angka):
    return sum(angka) #perhatian: tanda bintang tidak perlu dicantumkan kembali!

# Fungsi yang mengembalikan nilai boolean
def isEven(number):
    return number % 2 == 0
````

````python
# Uji fungsi
print(greet())

# Uji luasLingkaran
print(luasLingkaran(7))    # 3.14 * 7^2 = 153.86

# Uji luasBalok
print(luasBalok(2, 3, 4))  # 2 * 3 * 4 = 24

# Uji jumlah dengan banyak input
print(jumlah(1, 2, 3, 4, 5))   # 15
print(jumlah(10, 20))          # 30

# Uji isEven
print(isEven(10))   # True, karena 10 habis dibagi 2
print(isEven(7))
````

## Descriptive Statistic

Konsep Dasar

Descriptive Statistics (Statistik Deskriptif) adalah cabang statistik yang berfungsi untuk:
* Mendeskripsikan data apa adanya.
* Menyajikan informasi tentang karakteristik utama dari data (tanpa membuat kesimpulan lebih lanjut).


ðŸ“Œ Beberapa ukuran penting dalam statistik deskriptif:

**Tendensi Sentral (Central Tendency)**
* Mean: nilai rata-rata.
* Median: nilai tengah.
* Mode: nilai yang paling sering muncul.

**Ukuran Penyebaran (Dispersion/Variability)**
* Range: selisih nilai maksimum & minimum.
* Variance: seberapa jauh data tersebar dari rata-rata.
* Standard Deviation (SD): akar kuadrat variance, menunjukkan seberapa besar penyimpangan data dari rata-rata.

**Distribusi Data**
* Membagi data ke dalam interval/kategori.
* Digunakan untuk melihat pola sebaran nilai.

````python
import pandas as pd
import numpy as np

data = {
    "Nama": ["Ayu", "Budi", "Citra", "Deni", "Eka", "Fajar", "Gita", "Hadi", "Indah", "Joko"],
    "Nilai_Ujian": [80, 75, 90, 60, 85, 70, 95, 88, 77, 65]
}

df = pd.DataFrame(data)
print(df)
````

````python
#Tendensi Sentral (Central Tendency)

print("Mean (Rata-rata):", df["Nilai_Ujian"].mean())
print("Median (Nilai tengah):", df["Nilai_Ujian"].median())
print("Mode (Nilai sering muncul):", df["Nilai_Ujian"].mode()[0])
````

````python
#Ukuran Penyebaran (Dispersion/Variability)

print("Minimum:", df["Nilai_Ujian"].min())
print("Maksimum:", df["Nilai_Ujian"].max())
print("Range:", df["Nilai_Ujian"].max() - df["Nilai_Ujian"].min())
print("Variance:", df["Nilai_Ujian"].var())
print("Standard Deviation:", df["Nilai_Ujian"].std())
````

````python
bins = [50, 60, 70, 80, 90, 100]
kategori = pd.cut(df["Nilai_Ujian"], bins)
print(df.groupby(kategori).size())
````

````python
import matplotlib.pyplot as plt

plt.hist(df["Nilai_Ujian"], bins=5, edgecolor="black")
plt.title("Distribusi Nilai Ujian")
plt.xlabel("Nilai")
plt.ylabel("Frekuensi")
plt.show()
````

## Inferential Statistic

Konsep Dasar

Statistik Inferensial adalah cabang statistik yang digunakan untuk:
* Membuat kesimpulan tentang populasi berdasarkan sampel.
* Melakukan pengujian hipotesis untuk mengetahui apakah suatu temuan signifikan atau hanya kebetulan.
* Mengukur hubungan antar variabel.

Beberapa teknik penting yang sering digunakan:

1. t-test
  
  Menguji perbedaan rata-rata.
  * One-sample t-test: apakah rata-rata satu sampel berbeda dari nilai tertentu.
  * Independent t-test: apakah dua kelompok independen punya rata-rata berbeda.
  * Paired t-test: apakah ada perbedaan rata-rata dari dua kondisi yang berpasangan (misal pre-test & post-test).

2. Chi-Square Test

  Menguji apakah ada hubungan antara dua variabel kategorik.

3. ANOVA (Analysis of Variance)

  Menguji apakah ada perbedaan rata-rata lebih dari 2 kelompok.

4. Korelasi

  Pearson Correlation: mengukur hubungan linear antar dua variabel numerik.

  Spearman Correlation: mengukur hubungan monotonik (berdasarkan ranking, tidak harus linear).

````python
# One-sample t-test
# Kasus: apakah rata-rata nilai mahasiswa berbeda dari 75?

import numpy as np
from scipy import stats

nilai = [80, 75, 90, 60, 85, 70, 95, 88, 77, 65]

t_stat, p_val = stats.ttest_1samp(nilai, 75)
print("t-stat:", t_stat, "p-value:", p_val)
````

````python
# Independent t-test
# Kasus: apakah rata-rata nilai kelas A dan kelas B berbeda?

kelas_A = [80, 75, 90, 85, 70]
kelas_B = [60, 65, 70, 72, 68]

t_stat, p_val = stats.ttest_ind(kelas_A, kelas_B)
print("t-stat:", t_stat, "p-value:", p_val)
````

````python
# Paired t-test
# Kasus: apakah ada peningkatan signifikan antara pre-test dan post-test?

pre_test = [60, 65, 70, 72, 68]
post_test = [70, 75, 80, 78, 74]

t_stat, p_val = stats.ttest_rel(pre_test, post_test)
print("t-stat:", t_stat, "p-value:", p_val)
````

````python
# Chi-Square Test
# Kasus: apakah jenis kelamin berhubungan dengan kelulusan?

import numpy as np
from scipy.stats import chi2_contingency

# Tabel kontingensi (baris = gender [L, P], kolom = status [Lulus, Tidak])
data = np.array([[30, 10], [25, 15]])

chi2, p, dof, exp = chi2_contingency(data)
print("Chi2:", chi2, "p-value:", p)
````

````python
# ANOVA
# Kasus: apakah rata-rata nilai kelas A, B, dan C berbeda signifikan?

kelas_A = [80, 75, 90, 85, 70]
kelas_B = [60, 65, 70, 72, 68]
kelas_C = [88, 85, 84, 90, 86]

f_stat, p_val = stats.f_oneway(kelas_A, kelas_B, kelas_C)
print("F-stat:", f_stat, "p-value:", p_val)
````

````python
# Korelasi Pearson
# Kasus: apakah jam belajar berhubungan secara linear dengan nilai ujian?

jam_belajar = [2, 3, 4, 5, 6, 7, 8]
nilai = [60, 65, 70, 72, 75, 80, 85]

corr, p_val = stats.pearsonr(jam_belajar, nilai)
print("Pearson correlation:", corr, "p-value:", p_val)
````
